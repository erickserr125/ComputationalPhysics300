```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
#Part 1

"""
One way we could find a way to deal with noisy data is by 
averaging out the delta values.For example, given $delta$x = 1,
we average the function f(x) and f(x+$delta$x)
Once we've recieved this average, we can map this average to both points
for smoother results:
"""

#SAMPLE CODE GIVEN IN LESSON 6
f = lambda x: x*np.sin(x)

start, end = 0, 3
npoints = 100
x = np.linspace(start, end, npoints)
y = f(x) + np.random.rand(npoints) - 0.5

plt.plot(x,y)
plt.plot(x,f(x))

plt.xlabel('x')
plt.ylabel('$f(x)vs y$')
plt.xlim([start,end])
plt.show()

    
```

```python
#Part 1
#MY IMPLEMENTATION
def average(y):
    sum_avg = 0
    y_new = np.zeros(1)
    for val in y:
        sum_avg += val
        sum_avg /= 2
        y_new = np.append(y_new, (sum_avg))
            
    return y_new

#Implement 
y_prev = y


#Use iterations to decide how you want to fit the curve
for i in range(5):
    y_new = average(y_prev)
    #Remove the first element of the function to fit with x
    y_new = np.delete(y_new,0)
    y_prev = y_new


plt.plot(x, y_new)
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('$f(x) vs y_new$')
plt.show()
```
