```python
"""
Erick Serrano 
Week 8
Homework
1) Generate a data with random noise
2) Choose different polynomial functions to fit the data
3) Discuss which polynomial function yields the best
fit and give your reasoning
"""

#1)Generate a data with random noise



#Sample code taken from section 8.3: General Curve Fitting
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np

#Created a custom function
f = lambda x,a,b,c: a+b*x+c*(x**2)

#define the paramters for the plot
x_min, x_max = 0,10
npoints = 100
a,b,c = 9,-2.5,3.4

x = np.linspace(x_min, x_max, npoints)
y = f(x,a,b,c) + np.random.rand(npoints) - 0.5



plt.plot(x,y)
plt.show()
```

```python
#2) Choose different polynomial functions to fit the data

fit_1 = np.polyfit(x, y, 3, full = True)
fit_2 = np.polyfit(x,y,5, full = True)
fit_3 = np.polyfit(x,y,2, full = True)
fit_4 = np.polyfit(x, y, 1, full = True)

y_1 = np.polyval(fit_1[0],x)
y_2 = np.polyval(fit_2[0],x)
y_3 = np.polyval(fit_3[0],x)
y_4 = np.polyval(fit_4[0],x)


plt.plot(x, f(x, a, b, c), label = 'ideal function')
#plt.plot(x,y_1, label = 'cubic fit')

#plt.plot(x,y_2, label = 'quintic fit')
plt.plot(x,y_3, label = 'Quadratic fit')
plt.plot(x,y_4, label = 'linear fit', color = 'black')

plt.xlim([0,5])
plt.ylim([0,100])
plt.legend(fontsize = 12)

plt.show()
```
