```python
from math import factorial
import numpy as np
import matplotlib.pyplot as plt

"""
Calculating Omega based on a given q
"""

def omega(N, q):
    omegaVals = np.empty(0)
    """
    N is given, and q is a list of values
    Iterate through each value of q to get 
    a list of omegas
    """
    for x in q:
        omegaVals = np.append(omegaVals, (factorial(x+N-1)//factorial(x)//factorial(N-1)))
    return omegaVals
    
#q = 100, therefore qa + qb = 100 at all times
q = np.arange(101)
q = q.reshape((q.shape[0],1))

#Write a code to calculate as a function of qa, when
#1) Na = 300, Nb = 200, q = 100
omegaA = omega(300, q)
omegaB = omega(200, np.flip(q)) #In reverse so qa+qb = q

omegaA = omegaA.reshape(omegaA.shape[0],1)
omegaB = omegaB.reshape(omegaB.shape[0],1)

omegaTotal = omegaA*omegaB
```

```python

plt.plot(q, omegaA, '.r')
plt.xlabel('$q_a$')
plt.ylabel('$\Omega_A$')

plt.show()
```

```python
plt.plot(q, omegaB, '.r')
plt.xlabel('$q_b$')
plt.ylabel('$\Omega_B$')

plt.show()
```

```python
plt.plot(q, omegaTotal, '.r')
plt.xlabel('$q_total$')
plt.ylabel('$\Omega$')
plt.show()
```

```python
#Thus, N increases significantly between 40 and 60
```

```python
#Now we do this for Na = 3000, Nb = 2000
omegaAL = omega(3000, q)
omegaBL = omega(2000, np.flip(q)) #In reverse so qa+qb = q

omegaAL = omegaAL.reshape(omegaAL.shape[0],1)
omegaBL = omegaBL.reshape(omegaBL.shape[0],1)

omegaTotalL = omegaAL*omegaBL
```

```python
plt.plot(q, omegaTotalL, '.r')
plt.xlabel('$q_total$')
plt.ylabel('$\Omega_L$')
plt.show()
```
